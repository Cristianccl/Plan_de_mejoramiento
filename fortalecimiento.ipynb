```python
import random
import time

NUM_CASILLAS = 44

CASILLAS_ESPECIALES = {
    4: "Modo automático activado",
    8: "Corto circuito",
    11: "Ajuste de torque",
    13: "Restablecimiento de energía",
    15: "Mantenimiento preventivo",
    17: "Falla de motor",
    21: "Actualización de software",
    25: "Sobrecarga eléctrica",
    28: "Ajuste de torque",
    30: "Falla mecánica crítica",
    33: "Corto circuito",
    35: "Actualización de software",
    41: "Optimización del sistema",
    43: "Sensor fuera de calibración",
    44: "Línea de producción activa (LLEGADA)"
}

def lanzar_dado():
    return random.randint(1, 6)

def efecto_casilla(jugador, jugadores, posiciones, pierde_turno, turno_actual):
    nombre = CASILLAS_ESPECIALES.get(posiciones[jugador])
    if not nombre:
        return False
    print(f"{nombre} activado")

    if nombre == "Corto circuito":
        posiciones[jugador] = 1
        print(f"{jugador} retrocede al inicio")
    elif nombre == "Sobrecarga eléctrica":
        pierde_turno[jugador] = True
        print(f"{jugador} pierde el siguiente turno")
    elif nombre == "Falla de motor":
        posiciones[jugador] = max(1, posiciones[jugador] - 3)
        print(f"{jugador} retrocede 3 casillas")
    elif nombre == "Ajuste de torque":
        posiciones[jugador] += 2
        print(f"{jugador} avanza 2 casillas adicionales")
    elif nombre == "Sensor fuera de calibración":
        posiciones[jugador] = 5
        print(f"{jugador} salta a la casilla 5")
    elif nombre == "Modo automático activado":
        print(f"{jugador} obtiene un turno extra inmediato")
        return "extra_turno"
    elif nombre == "Mantenimiento preventivo":
        for j in jugadores:
            posiciones[j] = max(1, posiciones[j] - 1)
        print("Todos los jugadores retroceden 1 casilla")
    elif nombre == "Actualización de software":
        posiciones[jugador] = min(NUM_CASILLAS, posiciones[jugador] + 1)
        print(f"{jugador} avanza a la siguiente casilla por actualización")
    elif nombre == "Optimización del sistema":
        decision = input("¿Deseas avanzar 3 (A) o retroceder 2 (R)? ").strip().upper()
        if decision == "A":
            posiciones[jugador] += 3
        else:
            posiciones[jugador] = max(1, posiciones[jugador] - 2)
        print(f"{jugador} ha elegido su movimiento")
    elif nombre == "Falla mecánica crítica":
        posiciones[jugador] = 15
        print(f"{jugador} retrocede a la casilla de mantenimiento")
    elif nombre == "Restablecimiento de energía":
        posiciones[jugador] += 6
        print(f"{jugador} avanza 6 casillas")
        for j in jugadores:
            if j != jugador:
                posiciones[j] = max(1, posiciones[j] - 6)
                print(f"{j} retrocede 6 casillas por pérdida de energía")
    elif nombre == "Línea de producción activa (LLEGADA)":
        print(f"{jugador} ha llegado a la línea de producción y gana el juego")
        return "ganador"
    return False

def jugar_la_fabrica():
    print("BIENVENIDO A 'LA FÁBRICA'")
    num_jugadores = int(input("Ingrese el número de jugadores (mínimo 2): "))
    while num_jugadores < 2:
        num_jugadores = int(input("Debe haber al menos 2 jugadores. Intente nuevamente: "))

    jugadores = [f"Jugador {i+1}" for i in range(num_jugadores)]
    posiciones = {j: 1 for j in jugadores}
    pierde_turno = {j: False for j in jugadores}
    ganador = None
    turno = 0
    print("Comienza el juego")

    while not ganador:
        jugador = jugadores[turno % num_jugadores]
        if pierde_turno[jugador]:
            print(f"{jugador} pierde este turno por sobrecarga eléctrica")
            pierde_turno[jugador] = False
            turno += 1
            continue

        input(f"{jugador}, presiona ENTER para lanzar el dado...")
        dado = lanzar_dado()
        print(f"{jugador} sacó un {dado}")
        posiciones[jugador] += dado
        if posiciones[jugador] > NUM_CASILLAS:
            exceso = posiciones[jugador] - NUM_CASILLAS
            posiciones[jugador] = NUM_CASILLAS - exceso
            print(f"{jugador} se pasó y rebota {exceso} casillas atrás")
        print(f"{jugador} está ahora en la casilla {posiciones[jugador]}")

        if posiciones[jugador] in CASILLAS_ESPECIALES:
            resultado = efecto_casilla(jugador, jugadores, posiciones, pierde_turno, turno)
            if resultado == "ganador":
                ganador = jugador
                break
            elif resultado == "extra_turno":
                print(f"{jugador} repite turno inmediatamente")
                continue
        turno += 1
        time.sleep(1)

    print(f"FELICITACIONES, {ganador}, HAS GANADO EL JUEGO 'LA FÁBRICA'")

if __name__ == "__main__":
    jugar_la_fabrica()
```
