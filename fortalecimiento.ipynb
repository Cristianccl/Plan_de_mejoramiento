```python
import random

CASILLAS_TOTALES = 40

casillas_especiales = {
    5: "Corto circuito",
    8: "Sobrecarga eléctrica",
    11: "Falla de motor",
    15: "Ajuste de torque",
    18: "Sensor fuera de calibración",
    20: "Modo automático activado",
    23: "Mantenimiento preventivo",
    25: "Actualización de software",
    30: "Optimización del sistema",
    33: "Falla mecánica crítica",
    35: "Restablecimiento de energía",
    CASILLAS_TOTALES: "Línea de producción activa (LLEGADA)"
}

CASILLA_CALIBRAR_SENSOR = 22
CASILLA_OTRA_ACTUALIZACION = 28
CASILLA_MANTENIMIENTO_ANTERIOR = 23

def lanzar_dado():
    return random.randint(1, 6)

def mover_jugador(posicion, avance):
    nueva_posicion = posicion + avance
    if nueva_posicion > CASILLAS_TOTALES:
        exceso = nueva_posicion - CASILLAS_TOTALES
        nueva_posicion = CASILLAS_TOTALES - exceso
        print(f"Rebote, retrocedes {exceso} casillas.")
    elif nueva_posicion < 1:
        nueva_posicion = 1
    return nueva_posicion

def aplicar_efecto(jugador, jugadores, turno_actual):
    posicion = jugador["posicion"]

    if posicion not in casillas_especiales:
        return turno_actual

    efecto = casillas_especiales[posicion]
    print(f"Jugador {jugador['id']} cayó en '{efecto}'")

    if efecto == "Corto circuito":
        jugador["posicion"] = 1

    elif efecto == "Sobrecarga eléctrica":
        jugador["pierde_turno"] = True

    elif efecto == "Falla de motor":
        jugador["posicion"] = mover_jugador(posicion, -3)

    elif efecto == "Ajuste de torque":
        jugador["posicion"] = mover_jugador(posicion, 2)

    elif efecto == "Sensor fuera de calibración":
        jugador["posicion"] = CASILLA_CALIBRAR_SENSOR

    elif efecto == "Modo automático activado":
        print("Modo automático, juegas de nuevo inmediatamente.")
        turno_actual -= 1

    elif efecto == "Mantenimiento preventivo":
        for j in jugadores:
            j["posicion"] = mover_jugador(j["posicion"], -1)

    elif efecto == "Actualización de software":
        if posicion == 25:
            jugador["posicion"] = CASILLA_OTRA_ACTUALIZACION
        else:
            jugador["posicion"] = 25

    elif efecto == "Optimización del sistema":
        eleccion = ""
        while eleccion not in ["1", "2"]:
            eleccion = input("¿Deseas (1) avanzar 3 o (2) retroceder 2 casillas?: ")
        if eleccion == "1":
            jugador["posicion"] = mover_jugador(posicion, 3)
        else:
            jugador["posicion"] = mover_jugador(posicion, -2)

    elif efecto == "Falla mecánica crítica":
        jugador["posicion"] = CASILLA_MANTENIMIENTO_ANTERIOR

    elif efecto == "Restablecimiento de energía":
        jugador["posicion"] = mover_jugador(posicion, 6)
        for j in jugadores:
            if j != jugador:
                j["posicion"] = mover_jugador(j["posicion"], -6)

    elif "Línea de producción activa" in efecto:
        print(f"El jugador {jugador['id']} ha llegado a la LLEGADA y gana el juego.")
        return None

    return turno_actual

print("Bienvenido a 'La Fábrica'")

numero_jugadores = int(input("¿Cuántos jugadores participarán? (mínimo 2): "))
while numero_jugadores < 2:
    numero_jugadores = int(input("Debe haber al menos 2 jugadores: "))

jugadores = []
for i in range(numero_jugadores):
    jugadores.append({"id": i + 1, "posicion": 1, "pierde_turno": False})

ganador = False
turno = 0

while not ganador:
    jugador = jugadores[turno % numero_jugadores]

    if jugador["pierde_turno"]:
        print(f"El jugador {jugador['id']} pierde este turno por sobrecarga eléctrica.")
        jugador["pierde_turno"] = False
        turno += 1
        continue

    input(f"Jugador {jugador['id']}, presiona ENTER para lanzar el dado...")
    dado = lanzar_dado()
    print(f"Jugador {jugador['id']} sacó un {dado}")

    jugador["posicion"] = mover_jugador(jugador["posicion"], dado)
    print(f"Nueva posición del jugador {jugador['id']}: {jugador['posicion']}")

    nuevo_turno = aplicar_efecto(jugador, jugadores, turno)
    if nuevo_turno is None:
        ganador = True
        break

    print("Estado actual de los jugadores:")
    for j in jugadores:
        print(f"Jugador {j['id']}: Casilla {j['posicion']}")

    turno = nuevo_turno + 1

print("Fin del juego. Gracias por jugar 'La Fábrica'.")


```
